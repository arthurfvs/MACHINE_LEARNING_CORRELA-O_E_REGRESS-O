```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv")

# Separação de treino e teste
treino = df.iloc[0:126]
teste = df.iloc[126:138]

# Regressão
Y_treino = treino.iloc[:,[2]]  # PIB
X = ['BRL']              # Variável a ser "analisada"
X_treino = treino.loc[:,X]
mod = LinearRegression()
mod.fit(X_treino, Y_treino)
# Análise de R2
print(mod.score(X_treino, Y_treino))
```

```python
## Calculando a correlação entre o PIB e as variáveis de trafégo BRL, BRP, BRT e suas correspondentes de SP, RJ e PR.

# Include 'PIB' in the list of variables
variaveis = ['BRL','BRP','BRT','SLP','SPP','SPT','PRL','PRP','PRT','RJL','RJP','RJT', 'PIB']
df_corr = df[variaveis]
df_corr.corr()

correlacao_pib = df_corr.corr()['PIB']
print(correlacao_pib)
```

```python
##Executando a regressão linear entre PIB e BRL. Resultado de R2 e posteriormente, com as 12 variáveis.

Y_treino = treino.iloc[:,[2]]
X = ['BRL']              ## variavel que sera analisada
X_treino = treino.loc[:,X]

print(Y_treino , X_treino)

mod = LinearRegression()
mod.fit(X_treino, Y_treino)

print(mod.score(X_treino, Y_treino))

## ou

Y_treino = treino.iloc[:,[2]]
X = ['BRL','BRP','BRT','SLP','SPP','SPT','PRL','PRP','PRT','RJL','RJP','RJT']
X_treino = treino.loc[:,X]

print(Y_treino , X_treino)

mod = LinearRegression()
mod.fit(X_treino, Y_treino)

print(mod.score(X_treino, Y_treino))
```

```python
## 3 pares de variaveis para analisar

## SPT E RJT

Y_treino = treino.iloc[:,[2]]
X = ['SPT', 'RJT']            ## par de variavel escolhida para analisar os estados de SP e RJ
X_treino = treino.loc[:,X]

print(Y_treino , X_treino)

mod = LinearRegression()
mod.fit(X_treino, Y_treino)

print(mod.score(X_treino, Y_treino))

## PIB E BRL

Y_treino = treino.iloc[:,[2]]
X = ['PIB', 'BRL']            ## par de variavel escolhida para analisar moeda estatal e produto interno bruto
X_treino = treino.loc[:,X]

print(Y_treino , X_treino)

mod = LinearRegression()
mod.fit(X_treino, Y_treino)

print(mod.score(X_treino, Y_treino))

## ANO_MES E PIB

Y_treino = treino.iloc[:,[2]]
X = ['SPT', 'PIB']            ## par de variavel escolhida para analisar relação do estado de SP e produto interno bruto
X_treino = treino.loc[:,X]

print(Y_treino , X_treino)

mod = LinearRegression()
mod.fit(X_treino, Y_treino)

print(mod.score(X_treino, Y_treino))
```
